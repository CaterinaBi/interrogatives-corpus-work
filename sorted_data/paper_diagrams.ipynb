Visualisations for Baunaz & Bonan (2023)

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
from matplotlib.ticker import PercentFormatter
```

Figure 1: Ex situ vs in situ over time: raw data

```python
plt.rcParams["figure.figsize"] = [4.00, 5.00]
plt.rcParams["figure.autolayout"] = True

labels = ['ESLO 1', 'ESLO 2']

# data
ex_situ = [913, 481]
in_situ = [208, 780]

x = np.arange(len(labels))  # the label locations
width = 0.30  # the width of the bars

fig, ax = plt.subplots()
# ax.yaxis.set_major_formatter(mtick.PercentFormatter(5.0))
rects1 = ax.bar(x - width/2, ex_situ, width, color='paleturquoise', label='ex situ')
rects2 = ax.bar(x + width/2, in_situ, width, color='mediumturquoise', label='in situ')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total occurrences (raw numbers)')
ax.set_xlabel('Corpora')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)

plt.show()
```

Figure 2: Proportion of ex situ vs in situ over time

```python
plt.rcParams["figure.figsize"] = [4.00, 4.00]
plt.rcParams["figure.autolayout"] = True

labels = ['ESLO 1', 'ESLO 2']

# data
ex_situ = [81, 38]
in_situ = [19, 62]

x = np.arange(len(labels))  # the label locations
width = 0.30  # the width of the bars

fig, ax = plt.subplots()
# ax.yaxis.set_major_formatter(mtick.PercentFormatter(5.0))
rects1 = ax.bar(x - width/2, ex_situ, width, color='paleturquoise', label='ex situ')
rects2 = ax.bar(x + width/2, in_situ, width, color='mediumturquoise', label='in situ')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percentage of occurrences')
ax.set_xlabel('Corpora')
ax.set_xticks(x, labels)
ax.legend()

for p in rects1:
   height = p.get_height()
   ax.text(x=p.get_x() + p.get_width() / 2, y=height+.10,
      s="{}%".format(height),
      ha='center')

for p in rects2:
   height = p.get_height()
   ax.text(x=p.get_x() + p.get_width() / 2, y=height+.10,
      s="{}%".format(height),
      ha='center')

plt.show()
```
